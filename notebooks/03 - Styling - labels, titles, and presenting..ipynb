```python
from ggplot import *
import pandas as pd
import numpy as np
```

```python
ggplot(aes(x='cyl'), data=mtcars) +\
    geom_bar()
```

```python
ggplot(aes(x='cyl'), data=mtcars) +\
    geom_bar() +\
    xlab("Cylinders (cyl)") +\
    ylab("Count")
```

```python
ggplot(aes(x='cyl', y='mpg'), data=mtcars) +\
    geom_bar() +\
    xlab("Cylinders (cyl)") +\
    ylab("Count") +\
    ggtitle("Cylinder Distribution")
```
